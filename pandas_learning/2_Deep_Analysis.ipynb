```python
df = pd.DataFrame({
    'brand': ['Yum Yum', np.nan, 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', np.nan],
    'rating': [4, 4, 3.5, 87, 5]
})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(how = 'all') #it will delete only row or columns contains the Nan value
```

```python
df.dropna(how = 'any') #by default
```

```python
#to remove any duplicate items

df.drop_duplicates(subset = ['rating'])
```

```python
df.drop_duplicates(subset = ['rating'] , keep = 'first') #by default
```

```python
df.drop_duplicates(subset = ['rating'], keep = 'last')
```

```python
df.drop_duplicates(subset = ['rating'], keep = False)
```
